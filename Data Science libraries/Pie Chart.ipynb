**Create Pie chart with the help of Matplotlib library in Python**

```python
import matplotlib.pyplot as plt
```

```python
labels = 'Python', 'C/C++', 'Java', 'R', 'JavaScript'
Sizes = [35, 21, 20, 16, 8]
```

```python
fig = plt.figure(figsize=(10,7))
#Simple Pie chart without Lableing
plt.pie(Sizes)
plt.show()
```

```python
#Add labels
fig = plt.figure(figsize=(10,7))
plt.pie(Sizes, labels=labels)
plt.show()
```

```python
#Show the labels
fig = plt.figure(figsize=(10,7))
plt.pie(Sizes, labels=labels)
plt.show()
```

```python
fig = plt.figure(figsize=(10,7))
Explode= (0.23, 0, 0, 0.1, 0)
plt.pie(Sizes, labels=labels, autopct ="%1.1f%%", explode = Explode)
plt.show()
```

```python
fig = plt.figure(figsize=(10,7))
Explode= (0.23, 0, 0, 0, 0)
plt.pie(Sizes, labels=labels, explode = Explode, autopct ="%1.1f%%", shadow= True)
plt.show()
```

```python
fig = plt.figure(figsize=(10,7))
Explode= (0.23, 0, 0, 0.1, 0)
plt.pie(Sizes, labels=labels, explode = Explode, autopct ="%1.1f%%", shadow= True)
plt.show()
```

```python
fig = plt.figure(figsize=(10,7))
Explode= (0.23, 0, 0, 0.1, 0)
Colors = ['Lime', 'Blue', 'Yellow', 'Red', '#800080']
plt.pie(Sizes, labels=labels, explode = Explode, autopct ="%1.1f%%", shadow= True, colors = Colors)
plt.show()
```

```python
fig = plt.figure(figsize=(10,7))
Explode= (0.23, 0, 0, 0.1, 0)
Colors = ['Lime', 'Blue', 'Yellow', 'Red', '#800080']
plt.pie(Sizes, labels=labels, explode = Explode, autopct ="%1.1f%%", shadow= True, colors = Colors, startangle = 90)
plt.show()
```

```python
fig = plt.figure(figsize=(10,7))
Explode= (0.2, 0, 0, 0.1, 0)
Colors = ['Lime', 'Blue', 'Yellow', 'Red', '#800080']
plt.pie(Sizes, labels=labels, explode = Explode, autopct ="%1.1f%%", 
        shadow= True, colors = Colors, startangle = 90)

plt.legend(labels, loc='center right', bbox_to_anchor=(0.6, 0.3, 0.5, 0.5), 
           fontsize='large', title='Popularity', labelcolor='Blue')
plt.axis('equal')
plt.show()
```
