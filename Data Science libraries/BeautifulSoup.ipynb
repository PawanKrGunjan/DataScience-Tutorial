```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

from urllib.request import urlopen
from bs4 import BeautifulSoup
import re # for manupulating 
```

```python
url = "https://www.hubertiming.com/results/2018MLK"
```

```python
html=urlopen(url).read()
```

```python
soup = BeautifulSoup(html)
```

```python
title = soup.title
print(title)
print(title.text)
```

```python
links = soup.find_all("a")
links
```

```python
allrows = soup.find_all("tr")
for row in allrows:
    row_list = row.find_all("td")
print(row_list)
```

```python
for cell in row_list:
    print(cell.text)
    #print(cell)
```

```python
data = []
allrows = soup.find_all("tr")
for row in allrows:
    row_list = row.find_all("td")
    dataRow=[]
    for cell in row_list:
        #dataRow.append(cell)
        dataRow.append(cell.text)
    data.append(dataRow)
data = data[4:]
print(data[-2:])
```

```python
df2['ChipTime_minutes']=pd.to_timedelta(df2["('Chip Time',)"])
df2['ChipTime_minutes']=df2['ChipTime_minutes'].astype('timedelta64[s]')/60
print(df2[['Gender','ChipTime_minutes']].head())
#df2[['ChipTime_minutes']].info()
```

```python
plt.bar(df2['Gender'],df2['ChipTime_minutes'])
plt.xlabel('Gender')
plt.ylabel('ChipTime_Minutes')
plt.title("camparision of average minutes run by male and female")
```
